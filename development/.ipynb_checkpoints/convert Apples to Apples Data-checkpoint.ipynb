```python
import json
```

```python
with open('red.json') as file:
    red = json.load(file)
with open('green.json') as file:
    green = json.load(file)
```

```python
red
```

```python
green[0]["synonyms"]
```

```python
applesToApples = {"blackCards":[], "whiteCards":[]}
```

```python
for card in green:
    applesToApples["blackCards"].append({"text":card["word"] + " -- " + card["synonyms"], "pick":1})
```

```python
for card in red:
    applesToApples["whiteCards"].append(card["word"] + ": " + card["description"])
```

```python
applesToApples
```

```python
with open('a_to_a.json', 'w') as f:
     json.dump(applesToApples, f)
```
